```python
df.drop('currency', axis=1, inplace=True)
df.drop(['New', 'PendingS'], axis=1, inplace=True)
df['ship-country'].fillna('India', inplace=True)
df['ship-state'].fillna(method='ffill', inplace=True)
df['ship-city'].fillna(method='ffill', inplace=True)
df['ship-postal-code'].fillna(method='ffill', inplace=True)
df['fulfilled-by'].fillna('Merchant', inplace=True)
df.dropna(subset=['Amount'], inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
df['B2B'] = df['B2B'].astype(int)
df
```

```python
df.info()
```

## 1. Sales Overview

**Goal**: Understand overall performance, trends, and patterns over time.

🔸 **Actions:**

- Convert `'Date'` back to datetime for time-based analysis.

- Extract year/month/day if needed.

- Plot total sales over time.

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

daily_sales = df.groupby('Date')['Amount'].sum()

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
daily_sales.plot()
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. Product Analysis

**Goal**: Analyze category, size, quantity.

**🔸 Actions**:

- Count products sold by category

- Quantity distribution by size

```python
import seaborn as sns

# Category-wise product count
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Category'], y=df['Qty'], estimator=sum)
plt.title('Total Quantity Sold per Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Size-wise quantity distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Size', y='Qty', data=df)
plt.title('Quantity Distribution by Product Size')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3. Fulfillment Analysis

**Goal**: Explore fulfillment method impact

```python
fulfillment_sales = df.groupby('Fulfilment')['Amount'].sum().sort_values()

plt.figure(figsize=(8, 6))
fulfillment_sales.plot(kind='barh')
plt.title('Total Sales by Fulfilment Type')
plt.xlabel('Amount')
plt.tight_layout()
plt.show()
```

##  4. Customer Segmentation

- If you have more customer-level features (not present in this dataset), you could do clustering. For now:

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='B2B', y='Amount', data=df)
plt.title('Sales Amount for B2B vs Non-B2B')
plt.xticks([0, 1], ['Non-B2B', 'B2B'])
plt.tight_layout()
plt.show()
```

## 5. Geographical Analysis

```python

state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
state_sales.plot(kind='bar')
plt.title('Top 10 States by Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
